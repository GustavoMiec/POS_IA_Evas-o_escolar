```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#tabela
url = "https://raw.githubusercontent.com/pnferreira/fiap-ia-devs/main/dropout-inaugural.csv"
df = pd.read_csv(url)
```

```python
# Visualizar as primeiras linhas da base de dados
display(df.head())

# Visualizar as informações da base de dados
display(df.info())

# Ver os valores únicos da coluna Target
display(df['Target'].value_counts())
```

```python
# Tratar dados nulos
df.dropna(inplace=True)

# Tratar dados duplicados
df.drop_duplicates(inplace=True)

# Outras formas de pré-processamento
# Exemplo: Transformação de variáveis categóricas para numéricas
# Store the Target column before one-hot encoding
y = df['Target']
df = pd.get_dummies(df, drop_first=True)
```

```python
# Ensure 'Target' is not in the columns to be dropped if it still exists
columns_to_drop = [col for col in df.columns if col != 'Target']
# Normalização dos dados
scaler = StandardScaler()
X = df.drop(columns= [col for col in ['Target'] if col in df.columns], errors='ignore')
X_scaled = scaler.fit_transform(X)
```

```python
# Divisão entre treino e teste (70% treino, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)

# Treinamento do modelo (Random Forest)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Avaliação do modelo
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=['Dropout', 'Graduate'])
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
# Exibir métricas
print(f'Acurácia: {accuracy:.2f}')
print(f'Acurácia: {accuracy * 100:.2f}%') # Multiplica por 100 e adiciona o símbolo de porcentagem
print(report)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Dropout', 'Graduate'], yticklabels=['Dropout', 'Graduate'])
plt.xlabel('Predito')
plt.ylabel('Real')
plt.title('Matriz de Confusão')
plt.show()
```
